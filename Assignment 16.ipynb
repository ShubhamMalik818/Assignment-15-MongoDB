```python
1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?
```

```python
ANS- MongoDB is a document-oriented NoSQL database that is designed to store and manage unstructured data in a flexible and 
     scalable way. It uses a flexible data model based on BSON (Binary JSON) documents, which allows for a more natural representation 
     of complex data structures and relationships.

Non-relational databases, also known as NoSQL databases, are databases that do not use the traditional tabular schema used in relational databases. 
Instead, they use a variety of data models, such as key-value, document-oriented, column-family, or graph-based models, to store and manage data. 
NoSQL databases are often used in scenarios where the data is unstructured, dynamic, and evolving rapidly, and traditional relational databases 
are not well-suited to handle the complexity of the data.

MongoDB is preferred over SQL databases in scenarios where:

    1. Unstructured data: MongoDB is designed to handle unstructured or semi-structured data, such as JSON documents, where the schema is not fixed 
                          or predefined. This makes it easier to store and manage complex data structures and relationships, such as hierarchical 
                          or nested data.

    2. High scalability: MongoDB is highly scalable and can handle large amounts of data and high levels of traffic without sacrificing performance. 
                         This makes it well-suited for applications that need to scale rapidly and handle a large number of concurrent users.

    3. Flexible data model: MongoDB's flexible data model allows for easy data modeling and schema evolution. It supports dynamic schema changes, 
                            which means that you can add or remove fields to documents without having to modify the schema of the entire database.

    4. Developer productivity: MongoDB is document-oriented model is often preferred by developers because it allows them to work with data in a 
                               more natural and intuitive way. It also supports a rich query language and indexing capabilities, which makes it 
                               easier to retrieve and analyze data.

In summary, MongoDB is a popular NoSQL database that is well-suited for scenarios where the data is unstructured or evolving rapidly, 
and where scalability and flexibility are important. It is often preferred by developers because of its intuitive data model and 
rich query capabilities.
```

```python
2. State and Explain the features of MongoDB.
```

```python
ANS- MongoDB is a popular NoSQL database that stores data in a document-oriented manner. 
     Here are some of the key features of MongoDB:

    1. Document-oriented data model: MongoDB stores data in flexible, JSON-like documents that can have varying structures, 
                                      making it a good choice for handling unstructured or semi-structured data.

    2. Dynamic schema: Unlike traditional relational databases, MongoDB does not have a rigid schema, which means that you can change 
                       the schema of your data on the fly, without any downtime or migration.

    3. Scalability: MongoDB is designed to be horizontally scalable, allowing you to scale out by adding more servers to a cluster. 
                    This makes it a good choice for handling large amounts of data or high traffic loads.

    4. High availability: MongoDB provides built-in support for replication, which means that your data is automatically duplicated 
                          across multiple servers, providing high availability and data durability.

    5. Rich query language: MongoDB provides a powerful query language with support for advanced features like aggregation, indexing, 
                            and geospatial queries.

    6. Built-in sharding: MongoDB has built-in support for sharding, which allows you to partition your data across multiple servers, 
                          enabling you to handle even larger datasets.

    7. Flexible deployment options: MongoDB can be deployed on-premises or in the cloud, and supports a wide range of platforms and 
                                    programming languages.

    8. Community and ecosystem: MongoDB has a large and active community, with a rich ecosystem of third-party tools and libraries, 
                                making it easy to integrate with other systems and technologies.

In summary, MongoDB's features make it a powerful and flexible database system that can handle a wide range of use cases and data types.
```

```python
4. Using the database and the collection created in question number 3, write a code to insert one record, and insert many records. 
   Use the find() and find_one() methods to print the inserted record.
```

```python
import pymongo

# Replace with your MongoDB connection string
conn_str = "mongodb+srv://pwskills:pwskills@<cluster-url>"

# Create a MongoClient object
client = pymongo.MongoClient(conn_str)

# Select the "mydatabase" database
db = client["mydatabase"]

# Select the "customers" collection
col = db["customers"]

# Insert one record
record = {"name": "Nishant", "address": "Sector 23", "age": 23}
result = col.insert_one(record)
print("Inserted ID:", result.inserted_id)

# Insert many records
records = [
    {"name": "Bruno", "address": "Sector 23", "age": 2},
    {"name": "Srikar", "address": "Kurukshetra", "age": 23},
    {"name": "Sheru", "address": "Udaipur", "age": 24}
]
result = col.insert_many(records)
print("Inserted IDs:", result.inserted_ids)

# Find one record
result = col.find_one({"name": "Nishant"})
print("Found one record:", result)

# Find all records
results = col.find()
print("Found all records:")
for result in results:
    print(result)
    
    
    
In the above code, we first insert one record using the insert_one() method and then insert multiple records using the insert_many() method. 
We then use the find_one() method to find the record with name "Nishantt" and print it, and use the find() method to find all records 
and print them. 
Note that the find() method returns a cursor, which can be iterated over to retrieve all the documents in the collection.
```

```python
5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.
```

```python
ANS- The find() method is used to query the MongoDB database and retrieve documents that match certain criteria. 
     It takes one or more query filters as input and returns a cursor that can be used to iterate over the matching documents.

Here is how you can use the find() method to query the "customers" collection created in the previous example:
    
    import pymongo

# Replace with your MongoDB connection string
conn_str = "mongodb+srv://pwskills:pwskills@<cluster-url>"

# Create a MongoClient object
client = pymongo.MongoClient(conn_str)

# Select the "mydatabase" database
db = client["mydatabase"]

# Select the "customers" collection
col = db["customers"]

# Find all records where age is greater than or equal to 15
query = {"age": {"$gte": 15}}
results = col.find(query)

# Print the matching records
print("Matching records:")
for result in results:
    print(result)
    

In the above code, we first connect to the MongoDB database and select the "customers" collection. We then construct a query filter 
that matches all records where the "age" field is greater than or equal to 15. We pass this query filter to the find() method and iterate 
over the results using a for loop. 
Finally, we print each matching record to the console.

Note that the query filter is a dictionary that maps field names to matching criteria. In this example, we use the $gte operator to 
specify that we want to match records where the "age" field is greater than or equal to 15. MongoDB provides a wide range of query operators 
that can be used to construct more complex queries.    
```

```python
6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.
```

```python
ANS- The sort() method is used to sort the results of a MongoDB query in ascending or descending order. It takes one or more 
     sort keys as input and returns a cursor that can be used to iterate over the sorted documents.

Here is an example to demonstrate sorting in MongoDB:
    
    import pymongo

# Replace with your MongoDB connection string
conn_str = "mongodb+srv://pwskills:pwskills@<cluster-url>"

# Create a MongoClient object
client = pymongo.MongoClient(conn_str)

# Select the "mydatabase" database
db = client["mydatabase"]

# Select the "customers" collection
col = db["customers"]

# Find all records and sort by age in descending order
results = col.find().sort("age", pymongo.DESCENDING)

# Print the sorted records
print("Sorted records:")
for result in results:
    print(result)
    
    
In this example, we first connect to the MongoDB database and select the "customers" collection. We then use the find() method 
to retrieve all records in the collection and pass the result to the sort() method. We specify the "age" field as the sort key and 
use the pymongo.DESCENDING constant to indicate that we want to sort in descending order.

The sort() method returns a cursor that can be used to iterate over the sorted records. 
In this example, we use a for loop to print each record to the console.

Note that the sort() method can take multiple sort keys as input. 
For example, you can sort by age in descending order and then by name in ascending order by passing a list of sort keys to the sort() method:   
    
    results = col.find().sort([("age", pymongo.DESCENDING), ("name", pymongo.ASCENDING)])

```

```python
7. Explain why delete_one(), delete_many(), and drop() is used.
```

```python
ANS- In MongoDB, delete_one(), delete_many(), and drop() are used to remove documents and collections from a database.

    1. delete_one() is used to remove a single document from a collection that matches a specific query filter. 
       It takes one query filter as input and removes the first document that matches the filter.

    2. delete_many() is used to remove multiple documents from a collection that match a specific query filter. 
       It takes one query filter as input and removes all documents that match the filter.

Here is an example to demonstrate the use of delete_one() and delete_many():

    
    import pymongo

# Replace with your MongoDB connection string
conn_str = "mongodb+srv://pwskills:pwskills@<cluster-url>"

# Create a MongoClient object
client = pymongo.MongoClient(conn_str)

# Select the "mydatabase" database
db = client["mydatabase"]

# Select the "customers" collection
col = db["customers"]

# Delete one record where the name is "John"
query = {"name": "Shubham"}
result = col.delete_one(query)
print("Deleted document count:", result.deleted_count)

# Delete all records where age is greater than or equal to 15
query = {"age": {"$gte": 15}}
result = col.delete_many(query)
print("Deleted document count:", result.deleted_count)



In this example, we first connect to the MongoDB database and select the "customers" collection. We then use the delete_one() method 
to remove a single record where the "name" field is "Shubham" and the delete_many() method to remove all records where the "age" field 
is greater than or equal to 15. Both methods take a query filter as input to specify which documents to remove.

    3. drop() is used to remove an entire collection from a database. It takes no input and removes the entire collection and all of its documents.

Here is an example to demonstrate the use of drop():
    
    import pymongo

# Replace with your MongoDB connection string
conn_str = "mongodb+srv://pwskills:pwskills@<cluster-url>"

# Create a MongoClient object
client = pymongo.MongoClient(conn_str)

# Select the "mydatabase" database
db = client["mydatabase"]

# Select the "customers" collection and drop it
col = db["customers"]
col.drop()

In this example, we first connect to the MongoDB database and select the "customers" collection. We then use the drop() method to 
remove the entire collection and all of its documents. 
Note that this operation cannot be undone, so use it with caution.
```
